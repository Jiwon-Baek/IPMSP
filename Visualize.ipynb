```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os

dirpath = os.path.join(os.getcwd(),'output')
file = [0 for i in range(1)]
file[0] = os.path.join(dirpath, '240620_12_lr_0.0001_K_1_T_1_2_5_5\\log\\train_log.csv')
data = []
for i in file:
    data.append(pd.read_csv(i))
    
window_size = 20
fig = plt.figure(figsize=(18, 2 * len(file)))
axes = [fig.add_subplot(len(file), 3, i + 1) for i in range(len(file)*3)]
for i in range(len(file)):
    data[i]['MA'] = data[i][' reward'].rolling(window=window_size).mean()
    data[i]['MA_tard'] = data[i][' reward_tard'].rolling(window=window_size).mean()
    data[i]['MA_setup'] = data[i][' reward_setup'].rolling(window=window_size).mean()
    axes[3*i].plot(data[i].episode,data[i][' reward'])
    axes[3*i+1].plot(data[i].episode,data[i][' reward_tard'], c='red')
    axes[3*i+2].plot(data[i].episode,data[i][' reward_setup'], c='green')
    axes[3*i].plot(data[i].episode[window_size:],data[i]['MA'][window_size:], c='black')
    axes[3*i+1].plot(data[i].episode[window_size:],data[i]['MA_tard'][window_size:], c='black')
    axes[3*i+2].plot(data[i].episode[window_size:],data[i]['MA_setup'][window_size:], c='black')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os

filecode = 'F9_500_f7to1_lr_0.0001_K_1_T_1_2'


dirpath = os.path.join(os.getcwd(),'output')
filepath = os.path.join(dirpath, filecode)
file = [0 for i in range(7)]
code = ['0:10','5:5', '6:4', '7:3', '8:2','9:1','10:0']
for i, c in enumerate(code):
    file[i] = os.path.join(filepath,str(c)+'\\log\\train_log.csv')
data = []
for i in file:
    data.append(pd.read_csv(i))
    
window_size = 20
fig = plt.figure(figsize=(12, 14))
axes = [fig.add_subplot(7, 1, i + 1) for i in range(7)]
for i in range(7):
    data[i]['MA'] = data[i][' reward'].rolling(window=window_size).mean()
    axes[i].plot(data[i].episode,data[i][' reward'])
    axes[i].plot(data[i].episode[window_size:],data[i]['MA'][window_size:])
    axes[i].set_ylabel(code[i])
plt.suptitle(filecode)
plt.tight_layout()
plt.show()
```
